```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url= "https://drive.google.com/file/d/1wZtEdtQ4FCNA85Tjga9U5phmPDSkH4xA/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
path
```

```python
df.head(10)
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df.head(12)
```

```python
null_columns_percent=df.isnull().mean()*100
null_columns_percent
```

```python
df2=df.drop(columns=null_columns_percent[null_columns_percent>20].keys(),axis=1)
df2.head(12)
```

```python
df2.isnull().sum()
```

```python
df3_object=df2.select_dtypes(include=["object"])
df3_object.head()
```

```python
df3_obj_null_percent=df3_object.isnull().mean()*100
```

```python
null_object_lst=df3_obj_null_percent[df3_obj_null_percent>0].keys()
null_object_lst
```

```python
df3_object["MasVnrType"].mode()
```

```python
df3_object["MasVnrType"].fillna(df3_object["MasVnrType"].mode()[0])
```

```python
for i, var in enumerate(null_object_lst):
    df3_object[var].fillna(df3_object[var].mode()[0],inplace=True)
    print(var,"=",df3_object[var].mode()[0])
```

```python
df3_object.isnull().sum()
```

```python
plt.figure(figsize=(16,9))
for i,var in enumerate(null_object_lst):
    plt.subplot(4,3,i+1)
    
    plt.hist(df3_object[var],label=f"{var}")
    plt.hist(df[var].dropna(),label="Original")
    plt.legend()
    
```

```python
df.update(df3_object)
# df.drop(columns=df2.keys(),inplace=True)
```
