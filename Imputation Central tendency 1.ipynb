```python
df_null_percent=df.isnull().sum()/df.shape[0]*100
df_null_percent
```

```python
df2_drop_na=df.drop(columns=df_null_percent[df_null_percent>20].keys(),axis=1)
df2_drop_na.head()
```

```python
LotConfig_list=df2_drop_na.LotConfig.unique()
df2_drop_na.isnull().sum()
```

```python
df3_int_float=df2_drop_na.select_dtypes(include=["int64","float64"])
ext_col_list=[var for var in df3_int_float.columns if df3_int_float[var].isnull().sum()>0]
ext_col_list
```

```python
df[df.loc[:,"LotConfig"]=="Inside"]["LotFrontage"].replace(np.nan,df[df.loc[:,"LotConfig"]=="Inside"]["LotFrontage"].mean())
```

```python
df_copy=df.copy()
# for var in df_copy.LotConfig.unique():
#     df_copy.update(df[df.loc[:,"LotConfig"]==var]["LotFrontage"].replace(np.nan,df[df.loc[:,"LotConfig"]==var]["LotFrontage"].mean()))
```

```python
comp_var_list=["LotConfig","Exterior2nd","KitchenQual"]
for col,comp in zip(ext_col_list,comp_var_list):
    for var in df_copy[comp].unique():
        df_copy.update(df[df.loc[:,comp]==var][col].replace(np.nan,df[df.loc[:,comp]==var][col].mean()))
```

```python
df_copy[ext_col_list].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
for i, var in enumerate(ext_col_list):
    plt.subplot(3,1,i+1)
    sns.distplot(df[var],kde_kws={"color": "k", "lw": 3, "label": "Original"})
    sns.distplot(df_copy[var],kde_kws={"color": "r", "lw": 3, "label": f"{var}"})
```

```python
df_copy_median=df.copy()
for col,comp in zip(ext_col_list,comp_var_list):
    for var in df_copy[comp].unique():
        df_copy_median.update(df[df.loc[:,comp]==var][col].replace(np.nan,df[df.loc[:,comp]==var][col].median()))
```

```python
df_copy_median[ext_col_list].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
for i, var in enumerate(ext_col_list):
    plt.subplot(3,1,i+1)
    sns.distplot(df[var],kde_kws={"color": "k", "lw": 3, "label": "Original"},hist=False)
    sns.distplot(df_copy[var],kde_kws={"color": "r", "lw": 6, "label": f"mean"},hist=False)
    sns.distplot(df_copy[var],kde_kws={"color": "b", "lw": 3, "label": f"median"},hist=False)
```

```python
for i,var in enumerate(ext_col_list):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df_copy[var])
    plt.subplot(3,1,3)
    sns.boxplot(df_copy_median[var])
```
