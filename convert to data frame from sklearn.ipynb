```python
dif_MSZon=df.MSZoning.value_counts()-df3_null_rm_row.MSZoning.value_counts()
dif_MSZon
```

```python
MSzon_feature_col=pd.concat([df.MSZoning.value_counts()/df.shape[0]*100,df3_null_rm_row.MSZoning.value_counts()/
                             df3_null_rm_row.shape[0]*100
                           ,dif_MSZon/df.MSZoning.value_counts()*100],axis=1,keys=["df.MSZoning","df3_null_rm_row.MSZoning","percet_diff"])
MSzon_feature_col
```

```python
def feature_analyses_by_cols(var):
    df[var].value_counts()
    df3_null_rm_row[var].value_counts()
    dif=df[var].value_counts()-df3_null_rm_row[var].value_counts()
    return pd.concat([df[var].value_counts()/df.shape[0]*100,df3_null_rm_row[var].value_counts()/
                             df3_null_rm_row.shape[0]*100
                           ,dif/df[var].value_counts()*100],axis=1,keys=[f"df.{var}",f"df3_null_rm_row.var","percet_diff"])
feature_analyses_by_cols("Street")
```
