```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url= "https://drive.google.com/file/d/1wZtEdtQ4FCNA85Tjga9U5phmPDSkH4xA/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
path
```

```python
df.head()
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
null_colmns=df.isnull().sum()/df.shape[0]*100
df2_drop_na_col=df.drop(columns=null_colmns[null_colmns>20].keys())
df2_drop_na_col.isnull().sum()
```

```python
null_colmns
```

```python
df3_num=df2_drop_na_col.select_dtypes(include=["int64","float64"])
```

```python
df3_num.head()
df2_drop_na_col.info()
```

```python
df3_num.shape
```

```python
df3_num.isnull().any()
```

```python
df4_ext_null_col=[var for var in df3_num.columns if df3_num[var].isnull().sum() >0]
df4_ext_null_col
```

```python
plt.figure(figsize=(25,25))
for i,var in enumerate(df4_ext_null_col):
    plt.subplot(3,1,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={"color": "k", "lw": 3, "label": "EXTRACT NULL COLUMNS"})
    
```

```python
df5_mean=df3_num.fillna(df3_num.mean())
df5_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))
for i, var in enumerate(df4_ext_null_col):
    plt.subplot(3,1,i+1)
    sns.distplot(df3_num[var],kde_kws={"color": "k", "lw": 3, "label": "EXTRACT NULL COLUMNS"})
    sns.distplot(df5_mean[var],kde_kws={"color": "r", "lw": 3, "label": "Fill NULL COLUMNS by mean"})
```

```python
df3_num.isnull().sum()
```

```python
df6_median=df3_num.fillna(df3_num.median())
df6_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))
for i, var in enumerate(df4_ext_null_col):
    plt.subplot(3,1,i+1)
    sns.distplot(df3_num[var],kde_kws={"color": "k", "lw": 3, "label": "EXTRACT NULL COLUMNS"})
    sns.distplot(df6_median[var],kde_kws={"color": "r", "lw": 3, "label": "Fill NULL COLUMNS by Median"})
```

```python
df7_mode=df3_num.fillna(df3_num.mode())
df7_mode.isnull().sum().sum()
```

```python
plt.figure(figsize=(25,25))
for i, var in enumerate(df4_ext_null_col):
    plt.subplot(3,1,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={"color": "k", "lw": 3, "label": "EXTRACT NULL COLUMNS"})
    sns.distplot(df7_mode[var],bins=20,kde_kws={"color": "r", "lw": 3, "label": "Fill NULL COLUMNS by Mode"})
```

```python
plt.figure(figsize=(25,25))
for i, var in enumerate(df4_ext_null_col):
    plt.subplot(3,1,i+1)
    sns.distplot(df3_num[var],kde_kws={"color": "k", "lw": 6, "label": "EXTRACT NULL COLUMNS"})
    sns.distplot(df7_mode[var],kde_kws={"color": "r", "lw": 5, "label": "Fill NULL COLUMNS by Mode"})
    sns.distplot(df6_median[var],kde_kws={"color": "b", "lw": 3, "label": "Fill NULL COLUMNS by Median"})
    sns.distplot(df5_mean[var],kde_kws={"color": "g", "lw": 3, "label": "Fill NULL COLUMNS by Mean"})
```

```python

for i,var in enumerate(df4_ext_null_col):
    plt.figure(figsize=(10,15))
    plt.subplot(4,1,1)
    sns.boxplot(df[var])
    plt.subplot(4,1,2)
    sns.boxplot(df5_mean[var])
    plt.subplot(4,1,3)
    sns.boxplot(df6_median[var])
    plt.subplot(4,1,4)
    sns.boxplot(df7_mode[var])
```

```python
df_concat=pd.concat([df[df4_ext_null_col],df5_mean[df4_ext_null_col],df6_median[df4_ext_null_col],df7_mode[df4_ext_null_col]],axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
